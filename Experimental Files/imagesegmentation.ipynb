```python
import cv2
import os

# Define the directory paths
RAW_DATA_DIR = r"C:/Users/dunli/Downloads/DeepBacs_Data_Object_Detection_E.coli_Growth_Stage/image_patches/train_images"
BINARIZED_DATA_DIR = r"C:\Users\dunli\Documents\STSY-project-main\Training Data\Binarizeddata"

# List all files in the directory
file_list = os.listdir(RAW_DATA_DIR)

# Process each image
for filename in file_list:
    # Read the image
    img_path = os.path.join(RAW_DATA_DIR, filename)
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    
    # Apply Gaussian blur to reduce noise
    img_blur = cv2.GaussianBlur(img, (5, 5), 0)
    
    # Apply Otsu's thresholding
    _, binarized_img = cv2.threshold(img_blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    
    # Save the binarized image
    output_path = os.path.join(BINARIZED_DATA_DIR, f"{filename.split('.')[0]}_binarized.png")
    cv2.imwrite(output_path, binarized_img)
    
    print(f"{filename} binarized and saved to {output_path}")
```
